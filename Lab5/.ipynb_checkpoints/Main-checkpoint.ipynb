```python
import pandas as pd
import gensim
import nltk
import re
import numpy as np
from nltk.corpus import gutenberg
from string import punctuation
from nltk.collocations import TrigramCollocationFinder
from nltk.collocations import TrigramAssocMeasures
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('news.csv')
df
```

```python
total_nulls = df[df.text.str.strip() ==''].shape[0]
print("Порожні документи:", total_nulls)
```

```python
wpt = nltk.WordPunctTokenizer()
stop_words = nltk.corpus.stopwords.words('english')
```

```python
def preproc_doc(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text, re.I|re.A)
    text = text.lower()
    text = text.strip()
    tokens = wpt.tokenize(text)
    filtered_tokens = [token for token in tokens if token not in stop_words]
    text = ' '.join(filtered_tokens)
    return text
```

```python
preproc_corpus = np.vectorize(preproc_doc)
p_corpus = preproc_corpus(df['text'])
```

```python
# Токенізація по словам
tokenized_docs = []
for doc in p_corpus:
    tokens = nltk.word_tokenize(doc)
    tokenized_docs.append(tokens)

# Виведення результату
print(tokenized_docs)
```

```python
bigram = gensim.models.Phrases(tokenized_docs, min_count=20,threshold=20, delimiter='_')
bigram_model = gensim.models.phrases.Phraser(bigram)
```

```python
norm_corpus_bigrams = [bigram_model[doc] for doc in tokenized_docs]
dictionary = gensim.corpora.Dictionary(norm_corpus_bigrams)
```

```python
dictionary.filter_extremes(no_below=20, no_above=0.6)
```

```python
bow_corpus = [dictionary.doc2bow(text) for text in norm_corpus_bigrams]
```

```python
topics_num = [2, 5, 7, 10, 15, 20]
coherence_scores_cv = []
coherence_scores_mass = []
perplexity_scores = []
for num_topics in topics_num:
    lda_model = gensim.models.LdaModel( corpus=bow_corpus, id2word=dictionary, chunksize=1740, alpha='auto', eta='auto', random_state=0, iterations=500, num_topics=num_topics, passes=20, eval_every=None)
    cv_coherence_model_lda = gensim.models.CoherenceModel(model=lda_model, corpus=bow_corpus, texts=norm_corpus_bigrams, dictionary=dictionary, coherence='c_v')
    avg_coherence_cv = cv_coherence_model_lda.get_coherence()
    umass_coherence_model_lda = gensim.models.CoherenceModel(model=lda_model, corpus=bow_corpus, texts=norm_corpus_bigrams, dictionary=dictionary, coherence='u_mass')
    avg_coherence_umass = umass_coherence_model_lda.get_coherence()
    perplexity = lda_model.log_perplexity(bow_corpus)
    coherence_scores_cv.append(avg_coherence_cv)
    coherence_scores_mass.append(avg_coherence_umass)
    perplexity_scores.append(perplexity)
```

```python
coherence_scores_cv, coherence_scores_mass, perplexity_scores
```

```python
plt.plot(topics_num, coherence_scores_cv)
plt.xlabel("Num Topics")
plt.ylabel("Coherence Score (CV)")
plt.show()
```

```python
plt.plot(topics_num, coherence_scores_mass)
plt.xlabel("Num Topics")
plt.ylabel("Coherence Score (UMass)")
plt.show()
```

```python
plt.plot(topics_num, perplexity_scores)
plt.xlabel("Num Topics")
plt.ylabel("Perplexity score")
plt.show()
```

# 2.Використати текст chesterton-thursday.txt з корпусу gutenberg бібліотеки nltk та вивести ключові триграми.

```python
chesterton = gutenberg.sents('chesterton-thursday.txt')
remove_terms = punctuation + '0123456789'
sense = [ [word.lower() for word in sent if word not in
remove_terms] for sent in chesterton]
sense = [' '.join(tok_sent) for tok_sent in sense]
```

```python
sense = filter(None, preproc_corpus(sense))
```

```python
finder = TrigramCollocationFinder.from_documents([item.split() for item in sense])
trigram_measures = TrigramAssocMeasures()
```

```python
finder.nbest(trigram_measures.raw_freq, 10)
```

```python
finder.nbest(trigram_measures.pmi, 10)
```

```python
finder.nbest(trigram_measures.likelihood_ratio, 10)
```
